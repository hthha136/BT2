```python
df = pd.read_csv('OnlineRetail.csv', encoding="latin1")
df.head(3)
```

```python
df.info()
```

```python
# Tách cột InvoiceDate thành 3 cột Day, Month, Year

df['Day']=pd.to_datetime(df['InvoiceDate'],format='%m/%d/%Y %H:%M').dt.day
df['Month']=pd.to_datetime(df['InvoiceDate'],format='%m/%d/%Y %H:%M').dt.month
df['Year']=pd.to_datetime(df['InvoiceDate'],format='%m/%d/%Y %H:%M').dt.year
df['Hour']=pd.to_datetime(df['InvoiceDate'],format='%m/%d/%Y %H:%M').dt.hour
df['Minute']=pd.to_datetime(df['InvoiceDate'],format='%m/%d/%Y %H:%M').dt.minute
df.head()
```

```python
conditions = [(df['Month'] >= 1) & (df['Month'] <= 3), (df['Month'] >= 4) & (df['Month'] <= 6), (df['Month'] >= 7) & (df['Month'] <= 9), (df['Month'] >= 10) & (df['Month'] <= 12)]
choices = [1,2,3,4]
df['Previous'] = np.select(conditions, choices, default=0)
df.head()
```

```python
df['Amount'] = df['Quantity'] * df['UnitPrice']
df.head()
```

```python
conditions = [(df['Country'] == 'United Kingdom') & (df['Previous'] == 4), (df['Country'] == 'France')]
choices = ['10%', '5%']
c = [0.9,0.95]
df['Discount'] = np.select(conditions, choices, default='0%')
df['Pay'] = np.select(conditions, c, default=1)
df.head()
```

```python
df['Total'] = df['Amount'] * df['Pay']
df.head()
```
